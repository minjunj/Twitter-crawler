Contents - 문장(내용)에서 명사 추출하기
-------------

```python
okt_contents_pos=[]  #형태소와 품사 부분 저장을 위한 리스트

for data in data_contents: 
    #형태소와 품사 부분 리스트에 추가 
    #okt.pos()는 리스트 안에 (튜플)을 넣어서 리턴
    #okt_pos는 리스트이고,okt.pos() 결과도 리스트이므로 2차원리스트 생성 
    
    okt_contents_pos.append(okt.pos(data))
```

```python
#print(okt_contents_pos)  
```

```python
import numpy as np
print('● 데이터 타입 : ', type(okt_contents_pos)) 
print("● [(형태소, 품사), ...(형태소, 품사)] 리스트 개수 : ", len(okt_contents_pos) )
```

```python
okt_contents_pos_result =[ ]

#바깥쪽 리스트 okt_pos에서 안쪽 리스트 pos에 해당하는 100개의 리스트를 반복
for pos in okt_contents_pos : 
    for i in pos :   #안쪽 리스트 pos에서 각각의 튜플 i를 반복
        okt_contents_pos_result.append(i)  #(튜플) i 값을 추가하는 리스트 pos_result 
```

```python
print('● 데이터 타입 : ', type(okt_contents_pos_result)) 
print('● shape : ', np.shape(okt_contents_pos_result))
print("● 리스트에 있는 튜플 개수 : ", len(okt_contents_pos_result))     
```

```python
okt_contents_nouns=[]  #명사 부분 저장을 위한 리스트

for data in data_contents : 
    #명사 부분 리스트에 추가 
    #okt.nouns()는 리스트 안에 '스트링'을 넣어서 리턴
    #okt_nouns는 리스트이고,okt.nouns() 결과도 리스트이므로 2차원리스트 생성 
    
    okt_contents_nouns.append(okt.nouns(data))      
```

```python
#바깥쪽 리스트 안에 또 하나의 리스트가 포함되어 있으며
#내부 리스트 안에 튜플이 있는 구조
okt_contents_nouns
```

```python
print('● 데이터 타입 : ', type(okt_contents_nouns)) 
print("● ['스트링', ...'스트링'] 리스트 개수 : ", len(okt_contents_nouns) )
```

```python
okt_contents_nouns_result =[ ]

#바깥쪽 리스트 okt_nouns에서 안쪽 리스트 nouns에 해당하는 100개의 리스트를 반복
for nouns in okt_contents_nouns : 
    for i in nouns :   #안쪽 리스트 nouns에서 각각의 '스트링' i를 반복
        okt_contents_nouns_result.append(i)  #'스트링' i 값을 추가하는 리스트 nouns_result 
```

```python
print('● 데이터 타입 : ', type(okt_contents_nouns_result)) 
print('● shape : ', np.shape(okt_contents_nouns_result))
print("● 리스트에 있는 튜플 개수 : ", len(okt_contents_nouns_result))   
```

```python
from collections import Counter
```

```python
#명사 리스트 확인
print("명사 개수 : ", len(okt_contents_nouns_result))
```

```python
okt_contents_nouns_count = Counter(okt_contents_nouns_result)  

print('● 데이터 타입 : ',type(okt_contents_nouns_count))
print('● {명사 : 개수} ', okt_contents_nouns_count)
```

불용어
----------------------

```python
stop_word_contents = open("E:/ai/Mid/제출용/불용어사전.txt", encoding='utf-8').read()
print('● 불용어 사전 데이터 타입 : ',type(stop_word_contents))
print('● 불용어 사전 내용 ')  
print(stop_word_contents) #의미있는 결과를 보기 위해 추가하다보니 많아짐

# '이태원'은 무조건 불용어 처리를 해야함 어짜피 이태원 키워드를 중심으로 찾은 데이터기에 많을 수 밖에 없다.
#  이 데이터 분석의 핵심을 자명한 부분의 키워드를 제거하고 분석함으로서 더 의미있는 정보를 얻는 것이다.
#  이를 통해 그 당시 상황의 사람들이 어떤 생각을 하고 있었는지에 대해 좀 더 명확히 알 수 있다.
#  자명한 부분 : 사고 참사 사람 사망자 사망 압사 할로윈
```

```python
stop_okt_contents = [ i for i in okt_contents_nouns_result if i in stop_word_contents ] 
print("● 추출된 명사가 불용어에 해당되는 명사 : %d 개" %len(stop_okt_contents))
print(stop_okt_contents,'\n')
```

```python
non_stop_okt_contents = [ j for j in okt_contents_nouns_result if j not in stop_word_contents ] 
print("● 추출된 명사가 불용어에 해당되지 않는 명사 : %d 개" %len(non_stop_okt_contents))
print(non_stop_okt_contents)
```

글자 수 제한
---------

```python
print("● 불용어 사전을 통과한 명사 개수 : %d 개" %len(non_stop_okt_contents))
print("● 불용어 사전을 통과한 명사" )
print(non_stop_okt_contents)
```

```python
refine_non_stop_okt_contents = []

for i in non_stop_okt_contents :  
    #글자수가 2자 ~ 10자에 해당되는 단어만 추출
    if len(i) >= 2 and len(i) <= 10 : 
        refine_non_stop_okt_contents.append(i) 
        
print("● 글자 수가 2~10 사이 명사 : %d 개" %len(refine_non_stop_okt_contents))    
print(refine_non_stop_okt_contents)
```

다시 빈도수
---------

```python
re_okt_contents_nouns_count = Counter(refine_non_stop_okt_contents)  

#collections.Counter타입은 딕셔너리의 value를 기준으로 내림차순 정렬
print('● 데이터 타입 : ',type(re_okt_contents_nouns_count))
print('● {명사 : 개수} ',re_okt_contents_nouns_count)
```
