예) since:2020-01-01

until:년도-월-일

**4) 특정 계정에게 보낸 트윗만 보기**

그냥 그 단어쓰면 됨

**6) 특정 단어덩어리가 있는 트윗만 보기**
 
"단어"

filter:links

**5.2 adds의 경우**

**첫 요소에만 있는 답장 세기**

```python
# x 답글 , x리트윗, x마음
# x리트윗, x마음
# x 답글
# x 마음
# ... 등등의 다양한 경우의 수가 있는데 답글은 무조건 첫 요소이므로 첫 요소에 "답" 문자를 찾아 파싱
# 그후 숫자만 떼서 리스트화

chat = []

a = ""
b = []

for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    #print(i) 
    if (i[0].find("답") != -1): #첫 배열에 답글 값이 담겼나 확인
        #print(i)
        b = i[0].split(" ") #i[0]'x 답글'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
        chat.append(b[0])
    else:
        chat.append(0)
    #print(chat)
```

**첫 요소에 마음이 있는지 체크**

```python
#방금 경우의 수 처럼 마음은 1 2 3 번째에 다 존재할 수 있기에 하나하나 떼어서 분석
# 첫 번째에 없으면 일단 0으로 넣어줌
heart = []
for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    #print(i) 
    if (i[0].find("마") != -1): #첫 배열에 마음 값이 담겼나 확인
        #print(i)
        b = i[0].split(" ") #i[0]'x 마음'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
        heart.append(int(b[0]))
    else:
        heart.append(0)
    #print(heart)
```

**두 번째 요소에 마음이 있는지 체크**

```python
# 그 다음 두 번째에서 발견되면
#replace하는 방식으로 업데이트
n = 0 #정수로 몇 번째 트윗의 마음인지 파악힘
for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    
    #print(i) 
    if (len(i) == 2):
        #print(i) 
        if (i[1].find("마") != -1): #첫 배열에 마음 값이 담겼나 확인
            #print(i)
            b = i[1].split(" ") #i[0]'x 마음'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
            #print(b[1])
            del heart[n] #0이 든 부분을 원래 값이랑 교체
            heart.insert(n, int(b[1])) 
    n += 1
    #print(heart)
```

**세 번째 요소에 마음이 있는지 체크**

```python
# 두 번째와 같음


n = 0
for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    
    #print(i) 
    if (len(i) == 3):
        #print(i) 
        if (i[2].find("마") != -1): #세번째 배열에 마음 값이 담겼나 확인
            #print(i)
            b = i[2].split(" ") #i[0]'x 마음'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
            #print(b[1])
            del heart[n]
            heart.insert(n, int(b[1]))
        elif(i[1].find("마") != -1):
            b = i[1].split(" ") #i[0]'x 마음'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
            #print(b[1])
            del heart[n]
            heart.insert(n, int(b[1]))
    n += 1
    #print(heart)
```

**첫 요소에 리트윗이 있는지 체크**

```python
#마음 첫 번째와 동일
retweet = []

for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    #print(i) 
    if (i[0].find("리") != -1): #첫 배열에 리트윗 값이 담겼나 확인
        #print(i)
        b = i[0].split(" ") #i[0]'x 리트윗'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
        retweet.append(int(b[0]))
    else:
        retweet.append(0)
    #print(retweet)
```

**두 번쨰 요소에 리트윗이 있는지 체크**

```python
#마음 두 번째와 동일
n = 0
for i in add_list:
    i = i.split(",")# ,를 분기로 나눠서 str을 ㅣist화
    
    #print(i) 
    if (len(i) == 2):
        #print(i) 
        if (i[1].find("리") != -1): #첫 배열에 리트윗 값이 담겼나 확인
            #print(i)
            b = i[1].split(" ") #i[0]'x 리트윗'을 str로 가져옴 공백을 분기로 나눠서 숫자와 문자 분리
            #print(b[1])
            del retweet[n] #0이 든 부분을 원래 값이랑 교체
            retweet.insert(n, int(b[1])) 
    n += 1
    #print(retweet)
```
